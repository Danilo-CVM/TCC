```
#Tranformando o Banco de Dados para percentuais de retorno, passo necessário ao cálculo de beta.

retorno = dados_yahoo.pct_change()[1:]
retorno
```

```
#Definindo a função que Calcula o beta em relação à Ibovespa
# Fonte: https://blog.quantinsti.com/asset-beta-market-beta-python/

import statsmodels.api as sm
from statsmodels import regression

def linreg(x,y):
  x = sm.add_constant(x)
  model = regression.linear_model.OLS(y,x).fit()

  #removendo a constante
  x = x[:, 1]
  return model.params[0], model.params[1]


```
